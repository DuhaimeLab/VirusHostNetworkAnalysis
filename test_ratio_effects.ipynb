```python
# filter Aug4_predictions.tsv to only include 10 hosts and 100 viruses 
# read in the data from tsv to a pandas dataframe
import pandas as pd
df = pd.read_csv("Sample_Input/Sep29_predictions.tsv", sep="\t")
# the pairs column has virus:host pairs
# The unique viruses will make of the rows.
viruses = df['pairs'].str.split(':').str[0].unique()
hosts = df['pairs'].str.split(':').str[1].unique()
# pick 10 viruses from the list
viruses = viruses[:4000]
# pick 10 hosts from the list
hosts = hosts[:10]
# filter the dataframe to only include the selected viruses and hosts
df = df[df['pairs'].str.split(':').str[0].isin(viruses) & df['pairs'].str.split(':').str[1].isin(hosts)]
# save the filtered dataframe to a new tsv file
df.to_csv("Sample_Input/ratio_10x4000_sep29.tsv", sep="\t", index=False)

```

### Using ER Model

```python
ratio_1_1_er = ER(ratio_1_1, 0.1)
ratio_1_1_er.fill_ER_graph()
ratio_1_2_er = ER(ratio_1_2, 0.1)
ratio_1_2_er.fill_ER_graph()
ratio_1_10_er = ER(ratio_1_10, 0.1)
ratio_1_10_er.fill_ER_graph()
ratio_1_50_er = ER(ratio_1_50, 0.1)
ratio_1_50_er.fill_ER_graph()
ratio_1_100_er = ER(ratio_1_100, 0.1)
ratio_1_100_er.fill_ER_graph()
ratio_1_200_er = ER(ratio_1_200, 0.1)
ratio_1_200_er.fill_ER_graph()
ratio_1_300_er = ER(ratio_1_300, 0.1)
ratio_1_300_er.fill_ER_graph()

ratio_1_1_properties = BipartiteGraph(ratio_1_1_er)
ratio_1_2_properties = BipartiteGraph(ratio_1_2_er)
ratio_1_10_properties = BipartiteGraph(ratio_1_10_er)
ratio_1_50_properties = BipartiteGraph(ratio_1_50_er)
ratio_1_100_properties = BipartiteGraph(ratio_1_100_er)
ratio_1_200_properties = BipartiteGraph(ratio_1_200_er)
ratio_1_300_properties = BipartiteGraph(ratio_1_300_er)

title = "ER"
```

```python
#eigenvector centrality
ratio_1_1_properties.calculate_centrality(algorithm="eigenvector")
ratio_1_2_properties.calculate_centrality(algorithm="eigenvector")
ratio_1_10_properties.calculate_centrality(algorithm="eigenvector")
ratio_1_50_properties.calculate_centrality(algorithm="eigenvector")
ratio_1_100_properties.calculate_centrality(algorithm="eigenvector")
ratio_1_200_properties.calculate_centrality(algorithm="eigenvector")
ratio_1_300_properties.calculate_centrality(algorithm="eigenvector")
```

```python
# Make a boxplot with the eigenvector centrality values
import matplotlib.pyplot as plt
import seaborn as sns
nested_host = [list(ratio_1_1_properties.eigenvector_host.values()), list(ratio_1_2_properties.eigenvector_host.values()), 
               list(ratio_1_10_properties.eigenvector_host.values()), list(ratio_1_50_properties.eigenvector_host.values()), 
               list(ratio_1_100_properties.eigenvector_host.values()), list(ratio_1_200_properties.eigenvector_host.values()),
               list(ratio_1_300_properties.eigenvector_host.values())]

sns.boxplot(nested_host, color="lightblue", linecolor="black", linewidth=1)
plt.title("Eigenvector Centrality of Hosts " + title)
plt.ylabel("Eigenvector Centrality Distribution")
plt.xlabel("Number of Hosts and Viruses")
plt.xticks(range(len(nested_host)), ['10x10', '10x20', '10x100', '10x500', '10x1000', '10x2000', '10x3000'], rotation=45)
plt.show()
```
